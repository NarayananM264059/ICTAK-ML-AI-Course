```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt

df = pd.read_csv('data(1).csv')

df1=df
df
```

Check for null values

```python
df.isna().sum()
```

Fill the null values in calories with mean values

```python
x = df["Calories"].mean()

df["Calories"].fillna(x, inplace = True)
df
```

Convert the values in date to proper format

```python
df['Date'] = pd.to_datetime(df['Date'])
df
```

Check Values in'Date' and 'Duration'.Fill appropriate values

```python
for index, row in df.iterrows():
    if (row.Duration > 60) or (row.Duration < 30):
        df.at[index, 'Duration'] = 60
df
```

Fill null values in date column with appropriate value


```python
for index, row in df.iterrows():
    if pd.isna(row.Date):
        if (index != 0) and (index != df.shape[0]-1):
            prev_date = df.at[index - 1, 'Date']
            next_date = df.at[index + 1, 'Date']
            if (prev_date + pd.Timedelta(days=2)) == next_date:
                df.at[index, 'Date'] = prev_date + pd.Timedelta(days=1)
df
```

Removing Duplicates

```python
df.drop_duplicates(inplace = True)
df
```

Cleaned Data

```python
df.to_csv('cleaned_data.csv')
data = df[['Duration','Pulse','Maxpulse','Calories']]
data
```
