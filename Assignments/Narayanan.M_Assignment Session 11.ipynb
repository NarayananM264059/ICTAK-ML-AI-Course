Pre Processing Framingham Dataset

```python
df.isnull().sum()
```

```python
#Drop Education column
df=df.drop(['education'],axis=1)
df
```

```python
y=df['TenYearCHD']
X=df.drop(['TenYearCHD'],axis=1)
X.describe()
```

Model Train and Test

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2, random_state=1)
```

```python
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(missing_values=np.nan,strategy= 'mean')
X1_train= pd.DataFrame(imputer.fit_transform(X_train))
X1_train.columns = X_train.columns
X1_train.index = X_train.index
X1_test= pd.DataFrame(imputer.fit_transform(X_test))
X1_test.columns = X_test.columns
X1_test.index = X_test.index
```

```python
df.nunique()
```

```python
X_train['BPMeds']=X_train['BPMeds'].fillna(X_train['BPMeds'].mode().iloc[0])
X_train.isna().sum()
```

```python
X_test['BPMeds']=X_test['BPMeds'].fillna(X_test['BPMeds'].mode().iloc[0])
X_test.isna().sum()
```

```python
X_train.isnull().sum()
```

```python
X_train['cigsPerDay']=X_train['cigsPerDay'].fillna(X_train['cigsPerDay'].mean())
X_train['totChol']=X_train['totChol'].fillna(X_train['totChol'].mean())
X_train['BMI']=X_train['BMI'].fillna(X_train['BMI'].mean())
X_train['heartRate']=X_train['heartRate'].fillna(X_train['heartRate'].mean())
X_train['glucose']=X_train['glucose'].fillna(X_train['glucose'].mean())
X_train.isnull().sum()
```

```python
X_test['cigsPerDay']=X_test['cigsPerDay'].fillna(X_test['cigsPerDay'].mean())
X_test['totChol']=X_test['totChol'].fillna(X_test['totChol'].mean())
X_test['BMI']=X_test['BMI'].fillna(X_test['BMI'].mean())
X_test['heartRate']=X_test['heartRate'].fillna(X_test['heartRate'].mean())
X_test['glucose']=X_test['glucose'].fillna(X_test['glucose'].mean())
X_test.isnull().sum()
```

```python
X_train.describe()
```

```python
X_test.describe()
```

```python
from sklearn import preprocessing
standardisation=preprocessing.StandardScaler()
X_train=standardisation.fit_transform(X_train)
X_train=pd.DataFrame(X_train)
X_train.describe()
```

```python
X_test=standardisation.fit_transform(X_test)
X_test=pd.DataFrame(X_test)
X_test.describe()
```

```python
from sklearn.linear_model import LogisticRegression
log_reg = LogisticRegression()
log_reg.fit(X_train, y_train)
y_pred_log = log_reg.predict(X_test)
y_pred_log
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score,precision_score,recall_score
print('Accuracy: ' , accuracy_score(y_test,y_pred_log))
print('Precision: ', precision_score(y_test,y_pred_log))
print('Recall: ', recall_score(y_test,y_pred_log))
```

Date preprocessing

```python
y=df1['Admitted ']
X = df1.drop(['Admitted '], axis=1)
X.describe()
```

Split data into Train,Test

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2, random_state=1)
```

```python
from sklearn import linear_model
logr=linear_model.LogisticRegression()
logr.fit(X_train,y_train)
y_pred=logr.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score,precision_score,recall_score
print('Accuracy: ' , accuracy_score(y_test,y_pred))
print('Precision: ' , precision_score(y_test,y_pred))
print('Recall: ' , recall_score(y_test,y_pred))
```
